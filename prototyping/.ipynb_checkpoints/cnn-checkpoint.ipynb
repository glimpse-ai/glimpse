```python
import tensorflow as tf
import sys
import os
sys.path.append('..')
from glimpse.utils.layers import conv_1d, conv_2d, fully_connected
from glimpse.utils.pixnet import conv_block
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import h5py
```

```python
image_dims = [1250,640,3]
data_dir = '/media/marsdenlab/Data2/glimpse/data/dataset.hdf5'
data = h5py.File(data_dir, 'r')

def train():
  return get_dataset('train')


def val():
  return get_dataset('val')


def test():
  return get_dataset('test')


def get_dataset(setname):
  set = data.get(setname)

  if not set:
    raise BaseException('{} set does not exist yet in dataset'.format(setname))

  return set.get('images'), set.get('labels'), set.get('label_lens')
```

```python

X_train, Y_train, train_label_lens = train()
X_val, Y_val, val_label_lens = val()
X_test, Y_test, test_label_lens = test()

print X_train.shape
print Y_train.shape
print X_train[0].shape
print train_label_lens[0]
print np.amax(train_label_lens)
```

```python
plt.figure()
plt.imshow(X_train[0])
plt.colorbar()
plt.show()
plt.close()
```

```python
def repeat(x,axis,repeat):
    s = x.get_shape().as_list()
    splits = tf.split(value=x,num_or_size_splits=s[axis],axis=axis)
    rep = [s for s in splits for _ in range(repeat)]
    return tf.concat(rep,axis)

def resize_tensor(x,scales=[1,2,2,2,1]):
    out = x
    for i in range(1,len(scales)):
        out = repeat(out,i,scales[i])
    return out
```

```python
#network parameters
batch_size = 4
learning_rate = 1e-5
train_steps = 10000
print_step = 200
num_filters = 32
hidden_size = 2000
filter_dims=[5,5]

N = X_train.shape[0]
Nval = X_val.shape[0]
dml_length = Y_train.shape[1]-1
vocab_length = Y_train.shape[2]

def get_batch(X,Y,lab_len,N,n=32):
    inds = np.random.choice(range(N),size=n, replace=False)
    x = [X[i] for i in inds]
    y = [Y[i] for i in inds]
    lens = [lab_len[i] for i in inds]

    x = np.asarray(x)
    y = np.asarray(y)
    y = y[:,:dml_length,:]

    mask_mat = np.zeros((n,y.shape[1]))
    for i in range(n):
        l = lens[i]
        mask_mat[i,:l-1] = 1.0
    if len(x.shape) == 3:
        x = x.reshape((1,x.shape[0],x.shape[1],x.shape[2]))
        y = y.reshape((1,y.shape[0],y.shape[1]))
        y = y[:,:dml_length,:]
        mask_mat = np.zeros((1,y.shape[1]))
        mask_mat[0,:lens[0]-1] = 1.0
    
    return x,y,mask_mat

xb,yb,m = get_batch(X_train,Y_train,train_label_lens,N,batch_size)
print xb.shape
print yb.shape
print m.shape
print m[:,:20]
print m[:,-1]
print np.sum(m)
```

```python
#graph construction
x_image = tf.placeholder(shape=[batch_size]+image_dims,dtype=tf.float32)
y = tf.placeholder(shape=[batch_size,dml_length,vocab_length],dtype=tf.float32)
mask = tf.placeholder(shape=[batch_size,dml_length],dtype=tf.float32)
image_vec = conv_block(x_image,
                      num_filters=num_filters,
                      filter_dims=filter_dims,
                      fc_size=hidden_size,
                      batch_size=batch_size,
                      scope='conv_block')    
```

```python
act = tf.nn.relu
v = tf.reshape(image_vec,shape=[batch_size,hidden_size,1])
print v
v = conv_1d(v,dims=7,activation=act,filters=num_filters)
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
image_encoding = v
print v
v = resize_tensor(v,scales=[1,2,1])
print v
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
print v
v = resize_tensor(v,scales=[1,2,1])
print v
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
print v
intermediate_output = conv_1d(v,dims=7,filters=vocab_length)
print intermediate_output
```

```python
#second branch
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
#downsample
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters,strides=2)
print v

v = conv_1d(v,dims=7,activation=act,filters=4*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=4*num_filters)
#downsample
v = conv_1d(v,dims=7,activation=act,filters=4*num_filters,strides=2)
print v

#concatenate image vec again
v = tf.concat([v,image_encoding],axis=2)
print v

v = conv_1d(v,dims=7,activation=act,filters=4*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=4*num_filters)

print v
v = resize_tensor(v,scales=[1,2,1])
print v
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
print v
v = resize_tensor(v,scales=[1,2,1])
print v
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
v = conv_1d(v,dims=7,activation=act,filters=2*num_filters)
print v
final_output = conv_1d(v,dims=7,filters=vocab_length)
print final_output

prediction = tf.nn.softmax(final_output)
```

```python
l_intermediate = tf.nn.softmax_cross_entropy_with_logits(
    logits=intermediate_output,labels=y)*mask
l_final = tf.nn.softmax_cross_entropy_with_logits(logits=final_output,labels=y)*mask

loss = tf.reduce_mean(l_intermediate)+tf.reduce_mean(l_final)
```

```python
opt = tf.train.AdamOptimizer(learning_rate)
train = opt.minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
#Start the train loop
train_hist = []
val_hist = []

for i in range(train_steps):
    xb,yb,m = get_batch(X_train,Y_train,train_label_lens,N,n=batch_size)
    xb = xb/255
    
    l,_=sess.run([loss,train],{x_image:xb,y:yb,mask:m})

    if i%print_step == 0:
        print np.mean(xb)
        xb,yb,m = get_batch(X_val,Y_val,val_label_lens,Nval,n=batch_size)
        xb = xb/255
        lval=sess.run(loss,{x_image:xb,y:yb,mask:m})
        print "iter: {} Train: {}, Val: {}".format(i,l,lval)
        train_hist.append(l)
        val_hist.append(lval)
```

```python
plt.figure()
plt.plot(train_hist, linewidth=2,color='r',label='train')
plt.plot(val_hist, linewidth=2,color='g',label='val')
plt.legend()
plt.show()
```
